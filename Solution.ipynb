# Instructions

1. Create a new repository in the GitHub. Name it Iron+ your name.
	Named it IronAlfon

2. Clone that repo to your computer.

```bash
$ git clone https://github.com/Alfon22A/IronAlfon
```

3. Create a file in that folder. Name it about.txt and add a fun fact about yourself.

```bash
$ touch AboutAlfon.txt
$ echo "Hi! My name is Alfon and my favourite number is 22 because it's my birthday!" > AboutAlfon.txt
```

4. Add the file you just created to the git.

```bash
$ git add AboutAlfon.txt
```

5. Commit the changes.

```bash
$ git commit -m "About file"
```

6. Push the changes to the GitHub and check your repository to make sure everything is there.

```bash
$ git push origin main
```


# Pair Programming Exercise

1. Fork your classmate's repository.

2. Clone the repository, so that you have it locally and can make changes

```bash
$ git clone https://github.com/Alfon22A/IronNadin.git
```

3. Create a new branch named classmate.

```bash
$ git branch classmate
$ git checkout classmate #this one is to swap to the branch
```

4. Create a new file with your name and finish the sentence

```bash
$ echo "I enrolled Ironhack's Data Analytics bootcamp because I want to land a field-related job" > Alfon.txt
```
5. Add the changes to the git.

```bash
$ git add Alfon.txt
```

6. Commit and push the changes.

```bash
$ git commit -m "Alfon file"
$ git push origin classmate
```

7. Create a pull request to make changes appear in the original repository.

8. Merge the branch your classmate created.
